2. Теперь нас интересует столбец **"Пол, возраст"**. Сейчас он представлен в формате **<Пол , возраст , дата рождения >**. Например:

Создайте два новых признака **"Пол"** и **"Возраст"**. При этом важно учесть:
* Признак пола должен иметь 2 уникальных строковых значения: 'М' - мужчина, 'Ж' - женщина. 
* Признак возраста должен быть представлен целыми числами.

Выполните преобразование, ответьте на контрольные вопросы и удалите признак **"Пол, возраст"** из таблицы.

*Совет: обратите внимание на структуру текста в столбце, в части на то, как разделены параметры пола, возраста и даты рождения между собой - символом ' , '. 
Гарантируется, что структура одинакова для всех строк в таблице. Вы можете воспользоваться этим.*


```python
# ваш код здесь

# посмотрим на структуру первых 2х слов
col_old = "Пол, возраст"
col_new_1 = "Пол"
col_new_2 = "Возраст"
print(f"{col_old} до:")
display(df[col_old].apply(lambda x: "".join(x.split(" , ")[0])).unique())
display(df[col_old].apply(lambda x: "".join(x.split(" , ")[1])).unique())


# функция-преобразования
def map_sex(sex_age):
    mapping = {
        "Мужчина": "М",
        "Женщина": "Ж",
    }
    if pd.isna(sex_age):
        return np.nan
    level = sex_age.split()[0]
    return mapping.get(level, np.nan)


# функция-преобразования
def map_age(sex_age):
    if pd.isna(sex_age):
        return np.nan
    match = re.search(r"\d+", sex_age.split(" , ")[1])
    if match:
        return int(match.group())
    return np.nan


# Преобразование
df[col_new_1] = df[col_old].apply(map_sex)
df[col_new_2] = df[col_old].apply(map_age)

# Проверка
print(f"{col_new_1} после:")
display(df[col_new_1].unique())
print(f"{col_new_2} после:")
display(df[col_new_2].unique())

# Удаление столбца
df.drop(columns=[col_old], inplace=True)

# Проверка
print(f"Результат:")
display(df.head())
```

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
# ваш код здесь

# посмотрим на структуру первых 2х слов
col_old = "Опыт работы"
col_new = "Опыт работы (месяц)"
print(f"{col_old} до:")
display(df[col_old].apply(lambda x: " ".join(str(x).split()[:6])).unique())


# функция-преобразования
def exp_to_months(experience):
    if pd.isna(experience) or experience == "Не указано":
        return np.nan

    years = 0
    months = 0

    regex = r"(?:(\d+)\s*(?:лет|года|год))?\s*(?:и\s+)?\s*(?:(\d+)\s*(?:месяцев|месяц|мес))?"

    text = experience[:30].replace("Опыт работы ", "").strip()
    match = re.search(regex, text)
    if not match:
        raise ValueError("Совпадение не найдено")

    years = int(match.group(1)) if match.group(1) else 0
    months = int(match.group(2)) if match.group(2) else 0

    # в месяцы
    total_months = years * 12 + months
    return total_months


# Преобразование
df[col_new] = df[col_old].apply(exp_to_months)

# Проверка
print(f"{col_new} после:")
display(df[col_new].unique()[:5])

# Удаление столбца
df.drop(columns=[col_old], inplace=True)

# Проверка
print(f"Результат:")
display(df.head(7))
```

```python
# ваш код здесь

# посмотрим на структуру первых 2х слов
col_old = "Город, переезд, командировки"
col_new_1 = "Город"
col_new_2 = "Готовность к переезду"
col_new_3 = "Готовность к командировкам"
print(f"{col_old} до:")
display(df[col_old].apply(lambda x: x.split(",")[0].strip()).unique()[:10])
display(
    df[col_old]
    .apply(lambda x: x.split(",")[1].strip() if len(x.split(",")) > 1 else "")
    .unique()[:10]
)
display(
    df[col_old]
    .apply(lambda x: x.split(",")[2].strip() if len(x.split(",")) > 2 else "")
    .unique()[:10]
)


# функция-преобразования
def parse_city(destinations):
    city_part = re.split(r"[,(]", destinations)[0].strip()
    city = re.split(r"\s*,\s*м\.\s*", city_part)[0].strip()

    # Список городов
    city_central = ["Москва", "Санкт-Петербург"]
    city_million = [
        "Новосибирск",
        "Екатеринбург",
        "Нижний Новгород",
        "Казань",
        "Челябинск",
        "Омск",
        "Самара",
        "Ростов-на-Дону",
        "Уфа",
        "Красноярск",
        "Пермь",
        "Воронеж",
        "Волгоград",
    ]
    city_million_str = "город-миллионник"
    сity_other_str = "другие"

    # Разделение строки на части
    if city in city_central:
        return city
    elif city in city_million:
        return city_million_str
    else:
        return сity_other_str


# функция-преобразования
def parse_relocation(destinations):

    relocate_negative_pattern = r"\bне\s+готов(?:а)?\s+к\s+переезду\b"
    if re.search(relocate_negative_pattern, destinations, re.IGNORECASE):
        return False

    relocate_positive_patterns = [
        r"\bготов(?:а)?\s+к\s+переезду\b",
        r"\bхочу\s+переехать\b",
    ]
    for pattern in relocate_positive_patterns:
        if re.search(pattern, destinations, re.IGNORECASE):
            return True
    raise ValueError(f"Error in parse_relocation: {destinations}")


# функция-преобразования
def parse_travel(destinations):

    travel_negative_pattern = r"\bне\s+готов(?:а)?\s+к\s+командировкам\b"
    if re.search(travel_negative_pattern, destinations, re.IGNORECASE):
        return False

    travel_positive_patterns = [
        r"\bготов(?:а)?\s+к\s+командировкам\b",
        r"\bготов(?:а)?\s+к\s+редким\s+командировкам\b",
    ]
    for pattern in travel_positive_patterns:
        if re.search(pattern, destinations, re.IGNORECASE):
            return True
    return False


# Преобразование
df[col_new_1] = df[col_old].apply(parse_city)
df[col_new_2] = df[col_old].apply(parse_relocation)
df[col_new_3] = df[col_old].apply(parse_travel)

# Проверка
print(f"{col_new_1} после:")
display(df[col_new_1].unique())
print(f"{col_new_2} после:")
display(df[col_new_2].unique())
print(f"{col_new_3} после:")
display(df[col_new_3].unique())

# Удаление столбца
df.drop(columns=[col_old], inplace=True)

# Проверка
print(f"Результат:")
display(df.head())
```

2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.

Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
# ваш код здесь

delete_columns = [
    "Пропорция",
    "Курс Валюты",
    "Валюта ISO",
    "Курс на дату",
    "ЗП (руб)",
    "Сумма",
    "Валюта",
]
for col in delete_columns:
    if col in df.columns:
        df = df.drop(columns=[col])

display("------------------------------------------------------------------")
display("Проверка валюты и mapping-а описания")
display("--------------------------------------------")

display(
    df["ЗП"]
    .apply(lambda x: x.split()[1].strip() if len(x.split()) > 1 else "")
    .unique()
)

cur_mapping = {
    "Name in Data": [
        "руб.",
        "бел.руб.",
        "KZT",
        "EUR",
        "USD",
        "грн.",
        "сум",
        "KGS",
        "AZN",
    ],
    "Name ISO": [
        "RUB",
        "BYN",
        "KZT",
        "EUR",
        "USD",
        "UAH",
        "UZS",
        "KGS",
        "AZN",
    ],
    "proportion": [1, 1, 100, 1, 1, 10, 10000, 10, 1],
    "Name Full": [
        "рубль",
        "белорусский рубль",
        "казахстанский тенге",
        "евро",
        "доллар",
        "гривна",
        "узбекский сум",
        "киргизский сом",
        "азербайджанский манат",
    ],
}

df_cur_mapping = pd.DataFrame(cur_mapping)
display(df_cur_mapping)

display("------------------------------------------------------------------")
display("Парсинг ЗП и добавление ISO валюты из mapping-а")
display("--------------------------------------------")

df["Сумма"] = df["ЗП"].apply(lambda x: float(x.split()[0]))
df["Валюта"] = df["ЗП"].apply(
    lambda x: x.split()[1].strip() if len(x.split()) > 1 else ""
)

df = df.merge(df_cur_mapping, how="left", left_on="Валюта", right_on="Name in Data")
df = df.drop(columns=["Name in Data", "proportion", "Name Full"])
df.rename(columns={"Name ISO": "Валюта ISO"}, inplace=True)

display(df[["ЗП", "Сумма", "Валюта", "Валюта ISO"]].head())

display("------------------------------------------------------------------")
display("Загрузка курса валют")
display("--------------------------------------------")

df_currency = pd.read_csv("ExchangeRates.csv", sep=",")
df_currency

df["Обновление резюме"] = pd.to_datetime(df["Обновление резюме"], dayfirst=True)
df["Курс на дату"] = df["Обновление резюме"].dt.date
df_currency["date"] = pd.to_datetime(
    df_currency["date"], format="%d/%m/%y", errors="coerce"
).dt.date

# Проверка
display(df[["Обновление резюме", "Курс на дату"]].head())
display(df_currency.head(3))
display(df_currency.tail(3))

display("------------------------------------------------------------------")
display("Добавление курса валют")
display("--------------------------------------------")

df = df.merge(
    df_currency,
    how="left",
    left_on=["Валюта ISO", "Курс на дату"],
    right_on=["currency", "date"],
)
df = df.drop(columns=["date", "currency", "per", "time", "vol"])
df.rename(
    columns={"close": "Курс Валюты", "proportion": "Пропорция"},
    inplace=True,
)

# заполнение пропусков для RUB
display("Пропуски:")
display(df[df["Курс Валюты"].isna()]["Валюта ISO"].unique())
df["Пропорция"] = df["Пропорция"].fillna(1)
df["Курс Валюты"] = df["Курс Валюты"].fillna(1)

display(df[["ЗП", "Сумма", "Валюта", "Валюта ISO", "Курс Валюты", "Пропорция"]].head())


display("------------------------------------------------------------------")
display("Финальный расчет ЗП в рублях")
display("--------------------------------------------")

df["ЗП (руб)"] = df["Сумма"] * df["Курс Валюты"] / df["Пропорция"]

df.head()
```

6. В результате анализа мы обнаружили потенциальные выбросы в признаке **"Возраст"**. Это оказались резюме людей чересчур преклонного возраста для поиска работы. Попробуйте построить распределение признака в **логарифмическом масштабе**. Добавьте к графику линии, отображающие **среднее и границы интервала метода трех сигм**. Напомним, сделать это можно с помощью метода axvline. Например, для построение линии среднего будет иметь вид:

`histplot.axvline(log_age.mean(), color='k', lw=2)`

В какую сторону асимметрично логарифмическое распределение? Напишите об этом в комментарии к графику.
Найдите выбросы с помощью метода z-отклонения и удалите их из данных, используйте логарифмический масштаб. Давайте сделаем послабление на **1 сигму** (возьмите 4 сигмы) в **правую сторону**.

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
df = df.copy()

df["log_age"] = np.log(df["Возраст"])

plt.figure(figsize=(10, 6))
histplot = sns.histplot(df["log_age"], kde=True, bins=30, color="blue", alpha=0.7)

# std and mean
log_mean = df["log_age"].mean()
log_std = df["log_age"].std()


# lines
def line_value(num):
    return log_mean + num * log_std


histplot.axvline(log_mean, color="k", lw=2, label="Среднее")
histplot.axvline(
    line_value(3),
    color="r",
    linestyle="--",
    lw=2,
    label="3 сигмы (правая граница)",
)
histplot.axvline(
    line_value(-3),
    color="g",
    linestyle="--",
    lw=2,
    label="3 сигмы (левая граница)",
)
histplot.axvline(
    line_value(4),
    color="orange",
    linestyle="-",
    lw=2,
    label="4 сигмы (послабление)",
)
plt.legend()
plt.title("Распределение возраста в логарифмическом масштабе с асимметрией влево")
plt.xlabel("Логарифм возраста")
plt.ylabel("Количество")
plt.grid()
plt.show()

# z score
df["z_score"] = (df["log_age"] - log_mean) / log_std

# Поиск выбросов
outer_bounds = (df["z_score"] < -3) | (df["z_score"] > 4)

outliers = df[outer_bounds]

# Вывод выбросов
print("Выбросы в возрасте:")
display(outliers[["Возраст", "z_score"]])


before = df.shape[0]
print(f"\nДо удаления: {df.shape[0]}")

# Удаление выбросов
df = df[outer_bounds]

# Проверка
after = df.shape[0]
print(f"\nПосле удаления: {df.shape[0]}")
print(f"\nУдалено: {before - after}")
```
